Temukan akar dari:<br>

<center>f(x) = x^3 - 3x + 1 dalam interval: [0, 1]</center><br>

Syarat & Asumsi:<br>
- $f(x)$ adalah kontinu
- $f(0) = 1, \ f(1) = -1$
- $\Rightarrow f(a)f(b) < 0$

```python
import math

def f(x):
    # Definisi fungsi sesuai soal: f(x) = x^3 - 3x + 1
    return x**3 - 3*x + 1

def bisection_method(a, b, eps):
    # 1. CONVERGENCE ANALYSIS (Analisis Konvergensi)
    # Menghitung jumlah iterasi minimum n agar |x - r| <= eps
    # Menggunakan rumus n >= log2((b - a) / eps)
    n_required = math.ceil(math.log2((b - a) / eps))
    
    print(f"--- Convergence Analysis ---")
    print(f"Interval awal: [{a}, {b}]")
    print(f"Target error (epsilon): {eps}")
    print(f"Jumlah iterasi yang diperlukan (n): {n_required}\n")
    
    # 2. VALIDASI AWAL
    # Memastikan f(a) dan f(b) memiliki tanda berbeda
    u = f(a)
    v = f(b)
    if u * v >= 0:
        print("Metode Bisection gagal: f(a) dan f(b) harus memiliki tanda berbeda.")
        return None

    # 3. ALGORITMA BISECTION (Looping)
    print(f"{'Iterasi':<10} | {'Titik Tengah (c)':<15} | {'f(c)':<15} | {'Lebar Interval':<15}")
    print("-" * 65)
    
    for iteration in range(1, n_required + 1):
        c = (a + b) / 2
        w = f(c)
        interval_width = (b - a) / 2
        
        print(f"{iteration:<10} | {c:<15.6f} | {w:<15.6f} | {interval_width:<15.6f}")

        # Logika pembaruan interval (seperti di flowchart)
        if u * w < 0:
            b = c
            v = w
        else:
            a = c
            u = w
            
    return c

# Parameter Soal
a_start = 0
b_start = 1
epsilon = 0.02

# Jalankan Program
root = bisection_method(a_start, b_start, epsilon)

if root is not None:
    print(f"\nEstimasi akar terbaik: {root:.6f}")
```
